```python

# Carrega o arquivo XML
words_list = []
doc_nums = {}
total_doc = 1
for k in [4,5,6,7,8,9]:
    tree = ET.parse(f'C:/Users/ntana/Doutorado/BMT/CysticFibrosis2/data/cf7{k}.xml')
    root = tree.getroot()
    # Itera sobre todas as ocorrências da tag 'abstract'

    for (i , abstract), doc_num in zip(enumerate(root.findall('.//ABSTRACT')), root.findall('.//RECORDNUM')):
        print(f"Lendo documento {doc_num.text}")
        doc_nums[f'{i+total_doc-1}'] = doc_num.text
        if len(abstract.text)<10:
            warnings.warn(f'Abstract do documento {i+total_doc} está vazio', Warning)
        else:
            word_count = {}
            abstract_text = abstract.text
            abstract_text = re.sub(r'[.,:;()-]', '', abstract_text)
            abstract_text = re.sub(r'[\n]', ' ', abstract_text)
            abstract_text = re.sub(r'[\"]', '', abstract_text)
            abstract_text = abstract_text.lower()

            for word in abstract_text.split(' '):
                if len(word) >= 3:
                    try:
                        word_count[word] += 1
                    except:
                        word_count[word] = 1
            words_list.append(word_count)
            print("Criação do dicionário de palavras concluída\n")
    #break
    total_doc +=i+1 
print(f'Processo finalizado, foram lidos {total_doc} documentos')
```

```python
inv_list = {}
print("Iniciando gerador de lista invertida\n")
for i, documento, in enumerate(words_list):
    print(f"lendo palavras no documento {i+1}\n")
    for key in documento.keys():
        if key in inv_list.keys():
            print(f"atualizando lista invertida")
            #print(key)
            inv_list[f'{key}'].extend([i]*documento[f'{key}'])
        else:
            print(f"adicionando nova palavra na lista invertida")
            inv_list[f'{key}'] = [i]*documento[f'{key}']
    print(f"Finalizando análise do documento {i+1}\n")
print('Lista invertida finalizada')


```

```python
# Caminho do arquivo CSV
caminho_arquivo = 'inv_list.csv'

# Abre o arquivo CSV em modo de escrita
with open(caminho_arquivo, 'w', newline='') as arquivo_csv:
    # Define os campos do cabeçalho
    campos = ['palavra', 'incidencia']
    
    # Inicializa o escritor CSV
    escritor_csv = csv.DictWriter(arquivo_csv, fieldnames=campos, delimiter=';')
    
    # Escreve o cabeçalho
    escritor_csv.writeheader()
    
    # Escreve os dados do dicionário no arquivo CSV
    for chave, valor in inv_list.items():
        escritor_csv.writerow({'palavra': chave.upper(), 'incidencia': valor})

print("Dados do dicionário foram salvos em", caminho_arquivo)
```
